```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
import datetime
```

```python
df_train = pd.read_csv('train_users_2.csv')
df_test = pd.read_csv('test_users.csv')
labels = df_train["country_destination"]

def return_year(date):
    year = '20' + date[len(date)-2:len(date)]
    return year

Only_NDF = df_train[ df_train['country_destination'] == 'NDF' ] 
Non_NDF = df_train[ df_train['country_destination'] != 'NDF' ] 

NDF_Years = Only_NDF['date_account_created'].map( lambda x: return_year(str(x)) )
Non_NDF_Years = Non_NDF['date_account_created'].map( lambda x: return_year(str(x)) )

year_to_count_NDF = {}
for year in NDF_Years:

    
    if year not in year_to_count_NDF:
        year_to_count_NDF[year] = 1.0
    else:
        year_to_count_NDF[year] += 1.0

year_to_count_Non_NDF = {}
for year in Non_NDF_Years:
    
    if year not in year_to_count_Non_NDF:
        year_to_count_Non_NDF[year] = 1.0
    else:
        year_to_count_Non_NDF[year] += 1.0
        
for year in year_to_count_NDF:
    year_to_count_NDF[year] = year_to_count_NDF[year] / ( year_to_count_NDF[year] + year_to_count_Non_NDF[year]) 
    year_to_count_Non_NDF[year] = 1.0 - year_to_count_NDF[year]
    
    
    

Non_NDF_XY = []
for year in year_to_count_Non_NDF:
    Non_NDF_XY.append( (int(year),year_to_count_Non_NDF[year]) )

dtype = [('year', int), ('percentage', float)]
Non_NDF_XY = np.array(Non_NDF_XY,dtype = dtype)
Non_NDF_XY = np.sort(Non_NDF_XY,order = 'year')

Non_NDF_X = [entry[0] for entry in Non_NDF_XY ]
Non_NDF_Y = [entry[1] for entry in Non_NDF_XY ]




NDF_XY = []
for year in year_to_count_NDF:
    NDF_XY.append( (int(year),year_to_count_NDF[year]) )

NDF_XY = np.array(NDF_XY,dtype = dtype)
NDF_XY = np.sort(NDF_XY,order = 'year')

NDF_X = [entry[0] for entry in NDF_XY ]
NDF_Y = [entry[1] for entry in NDF_XY ]


```

```python
plt.plot(Non_NDF_X,Non_NDF_Y,label = 'Booking')
plt.plot(NDF_X,NDF_Y,'r-',label = 'No Booking')

plt.xlabel('Year')
plt.ylabel('% of Total Bookings')

plt.axis([2010, 2014, 0.0, 1.0])
plt.xticks( np.arange(2010,2015,1) )
plt.legend()
plt.show()
```
